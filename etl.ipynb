```python
import pandas as pd
```

```python
data = pd.read_excel(r"C:\Users\VitoriaRodrigues\Documents\Desafio\ETL\arquivos Pythom\BaseTargetListv.xlsx")
data.loc[:, 'BuyerSubPk'] = data['Buyer'] + data['BuyerSubSector']

data
```

### Número de vezes em que esse buyer foi considerado para deals neste sub-setor;

```python
df_buyers = pd.DataFrame(data) 

df_buyers= data[data['Deal Close Date'].notna()]

df_buyers.loc[:, 'BuyerSubPk'] = df_buyers['Buyer'] + df_buyers['BuyerSubSector']
df_buyers = df_buyers.groupby(['BuyerSubPk'])['Deal Close Date'].count().reset_index()

df_buyers
```

### Número de NDAs assinados por esse buyer para deals neste sub-setor; 

```python
df_buyers = pd.DataFrame(data) 

df_buyers.loc[:, 'BuyerSubPk'] = df_buyers['Buyer'] + '-' + df_buyers['BuyerSubSector']
df_buyers = df_buyers.groupby(['BuyerSubPk'])['SignedNDADate'].count().reset_index()

df_buyers
```

###	Número de No Gos do buyer neste sub-setor; 

```python
df_buyers = pd.DataFrame(data) 

df_buyers.loc[:, 'BuyerSubPk'] = df_buyers['Buyer'] + '-' + df_buyers['BuyerSubSector']
df_buyers = df_buyers.groupby(['BuyerSubPk'])['PassedReason'].count().reset_index()

df_buyers
```

### Número de No Gos, por razão de tamanho, neste sub-setor; 

```python
df_buyerf = pd.DataFrame(data)
df_buyerf = df_buyerf.groupby(['BuyerSubPk'])['PassedReason'].sum().reset_index()

df_size = df_buyerf['PassedReason'].isin(['Too Small', 'Too Big'])
df_filterd = df_buyerf[df_size]

df_filterd
```

### Número de NBOs enviadas pelo buyer para deals neste sub-setor; 

```python
df_client= pd.DataFrame(data) 

df_client = df_client.groupby(['ClientSubSector'])['SubmittedNBODate'].count().reset_index()
df_client
```

### Média do tamanho dos deals (considerando apenas deals com NDA assinado por esse buyer) para deals neste sub-setor; 

```python
df_buyers = pd.DataFrame(data)
pd.options.display.float_format = '{:.0f}'.format

df_buyers = df_buyers[df_buyers['EstimatedDealSize'].notnull()]   
deals_nda = df_buyers[df_buyers['SignedNDADate'].notnull()]

mean_size_deals= deals_nda.groupby('BuyerSubPk')['EstimatedDealSize'].mean().reset_index()

mean_size_deals
```


### Mediana do tamanho do deal (considerando apenas deals com NDA assinado por esse buyer) para deals neste sub-setor

```python
df_buyers = pd.DataFrame(data)
pd.options.display.float_format = '{:.0f}'.format

df_buyers = df_buyers[df_buyers['EstimatedDealSize'].notnull()]   
deals_nda = df_buyers[df_buyers['SignedNDADate'].notnull()]

media_size_deals= deals_nda.groupby('BuyerSubPk')['EstimatedDealSize'].median().reset_index()

media_size_deals
```

### Data da última NBO enviada pelo buyer para deals neste sub-setor;

### Número de deals fechados por esse buyer neste sub-setor; 

```python
df_client = pd.DataFrame(data) 
df_client= data[data['Deal Close Date'].notna()]

df_client = df_client.groupby(['Deal Close Date'])['ClientSubSector'].max().reset_index()
df_client
```

### Data do último deal fechado neste sub-setor;

```python
df_clients = pd.DataFrame(data)

df_clients = df_clients.groupby(['Deal Close Date'])['ClientSubSector'].max().reset_index()
df_clients
```

### Contato do buyer, considerando a data de contato mais recente;

```python
df_buyer_sub = pd.DataFrame(data)

df_buyer_sub = df_buyer_sub.astype({'Client': str, 'Buyer': str})
df_buyer_sub = df_buyer_sub.groupby('LastContactDate').agg({'Client': lambda x: ', '.join(x.unique()), 'Buyer': lambda x: ', '.join(x.unique())}).reset_index()

df_buyer_sub['LastContactDate'].iloc[-1]
df_buyer_sub
```
